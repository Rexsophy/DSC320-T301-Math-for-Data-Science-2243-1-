```python
def right_riemann_sum(func, a, b, n):
    """
    Calculates the right Riemann sum of a function over an interval.
    
    :param func: The mathematical function to integrate.
    :param a: The start of the interval.
    :param b: The end of the interval.
    :param n: The number of rectangles.
    :return: The right Riemann sum R_n.
    """
    delta_x = (b - a) / n
    total_sum = 0
    for k in range(1, n + 1):
        total_sum += func(a + k * delta_x) * delta_x
    return total_sum
```

```python
# Solution to 1a

import numpy as np
import matplotlib.pyplot as plt
import math

# Define the function
def f(x):
    return np.sqrt(4 - x**2)

# Create an array of x values from 0 to 2
x = np.linspace(0, 2, 400)

# Create an array of y values using the function
y = f(x)

# Plot the function
plt.plot(x, y, label=r'$f(x) = \sqrt{4 - x^2}$')

# Calculate the right Riemann sum with 4 rectangles
a, b, n = 0, 2, 4
delta_x = (b - a) / n
for i in range(1, n + 1):
    x_rect = a + i * delta_x  # Right edge of each rectangle
    y_rect = f(x_rect)
    rect = plt.Rectangle((x_rect - delta_x, 0), delta_x, y_rect, edgecolor='r', facecolor='none')
    plt.gca().add_patch(rect)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Right Riemann Sum of f(x) with n=4')
plt.legend()

# Show the plot
plt.show()

```
