#### Solving a System of Linear Equations

```python
# Solution for (1)
import numpy as np

# Define matrix A using the coefficients of the linear equations
A = np.array([[27, -10, 4, -29],
              [-16, 5, -2, 18],
              [-17, 4, -2, 20],
              [-7, 2, -1, 8]])

# Define vector b using the constants from the right side of the equations
b = np.array([1, -1, 0, 1])

# Use numpy's linear algebra solver to solve for x
x = np.linalg.solve(A, b)

print("Solution for x:", x)
```

```python
# Solution for 2(b)
import numpy as np
import matplotlib.pyplot as plt

# Define the range for the grid
x = np.arange(-10, 11, 1)
y = np.arange(-10, 11, 1)

# Create a meshgrid from the range
xx, yy = np.meshgrid(x, y)

# Flatten the grids and stack them into a 2xN array of points
grid_points = np.vstack([xx.flatten(), yy.flatten()])

# Define the transformation matrix T
T = np.array([[np.sqrt(2)/2, -np.sqrt(2)/2],
              [np.sqrt(2)/2, np.sqrt(2)/2]])

# Apply the transformation matrix to each point
transformed_points = np.dot(T, grid_points)

# Reshape the transformed points back into the grid format for plotting
xx_transformed, yy_transformed = transformed_points.reshape(2, xx.shape[0], xx.shape[1])

# Plot the original and transformed grids
plt.figure(figsize=(12, 6))

# Original grid
plt.subplot(1, 2, 1)
plt.scatter(xx, yy, c='red', label='Original grid')
plt.title('Original Grid')
plt.axis('equal') 

# Transformed grid
plt.subplot(1, 2, 2)
plt.scatter(xx_transformed, yy_transformed, c='blue', label='Transformed grid')
plt.title('Transformed Grid')
plt.axis('equal')  

# Show the plot with a legend
plt.legend()
plt.show()
```

The transformation matrix T associated with the linear transformation in question 2(b) rotates vectors in the plane. The matrix T represents a 45-degree rotation counterclockwise, as shown by the grid's orientation change from the original to the transformed version. This transformation preserves the length of vectors meaning no scaling or distortion occurs apart from the rotation. The grid spacing remains consistent post-transformation, signifying that distances and angles between points (apart from rotation) are maintained, which is a characteristic of orthogonal transformations.
